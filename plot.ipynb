```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/diabetes_model.csv")
df
```

```python
sns.scatterplot(x='BloodPressure',y='Age',data=df,hue='BMI')
plt.xlabel('bloodpressure')
plt.ylabel('Age')
plt.title('hormones')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(df['Pregnancies'],df['Glucose'])
plt.xlabel('pregnant')
plt.ylabel('Gllucose')
plt.xticks(rotation=90)
plt.show()
```

```python
df.plot(kind='scatter', x='Pregnancies', y='Glucose')
```

```python
df1=df.sort_values(by='Age')
df1=df1.head(10)
plt.bar(df1['Pregnancies'],df1['BloodPressure'])
plt.xticks(rotation=90)
```

```python
plt.plot(df1['Age'],df1['BloodPressure'],color='orange',marker='s',markersize='8')
plt.xticks(rotation=90)
plt.xlabel('students')
plt.ylabel('marks')
plt.ylim(50,200)
plt.title('studentmark')
plt.show()
```
